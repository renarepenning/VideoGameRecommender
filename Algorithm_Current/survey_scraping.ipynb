```python
from google.colab import auth
import pandas as pd 

auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('capstone_survey_responses').sheet1

rows = worksheet.get_all_values()

df = pd.DataFrame.from_records(rows)

# Read in all indie games, and sort into list
col_list = df[3].tolist()
col_list.pop(0)
indie_games = []

for entry in col_list:
  games = entry.split(", ")
  for game in games:
    indie_games.append(game)

for game in indie_games:
  if game == '' or game == 'none' or game == 'None' or game == 'n/a':
      indie_games.remove(game)


print(indie_games)

# Manually take out other extra info
indie_games = ['Super Auto Pets', 'Phigros', 'Monument Valley', 'Kenshi', 'Little Misfortune', 'The Gardens Between', 'Incremental Adventures', 'Bubble Spinner', 'Stardew Valley', 'Omori', 'Carto', 'Coc', 'Rust', 'Terraria', 'Hat In Time', 'Blind Forest', 'Enter the Gungeon', 'Among us', 'Dead by Daylight', 'Rusty Lake', 'Among Us', 'No Man\'s Sky', 'Musedash', 'Celeste', 'Hotline Miami', 'Heal Hitler', 'Hollow Knight', 'Magic the Gathering Arena', 'FTL', 'Binding of Isaac', 'Buried', 'The Imagined Leviathan', 'ZenGM', 'Shovel Knight', 'Super Meat Boy', 'Limbo', 'Shing!', 'Splitgate', 'Hacknet', 'Hollow Knight', 'Cube Escape', 'Magicka', 'Super Hot', 'Darkest Dungeon', 'Cup Head', 'Ori', 'Last of Us', 'Super Hexagon', 'Doki Doki Literature Club', 'Hotline Miami', 'Stardew Valley', 'Catan', 'For the King', 'Sunless Sea', 'Angry Birds', 'Rocket League', 'Super Meat Boy', 'Untitled Goose Game', 'Trine', 'Hades']
indie_games = set(indie_games)
data = {'Game': list(indie_games)}
df = pd.DataFrame(data)
print(df)
df.to_csv('Indie Games')
```
